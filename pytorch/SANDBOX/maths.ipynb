```python
import torch
from d2l import torch as d2l

x = torch.arange(5.0, requires_grad=True)
y = x * x
```

```python
y.backward(gradient=torch.tensor([1,1,1,1,1]), retain_graph=True)
x.grad
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

x = torch.arange(0, 3.1415 * 4, 0.1, requires_grad=True)
y = torch.sin(x)

y.sum().backward()

u = x.detach()
v = y.detach()
w = x.grad.detach()

plt.plot(u, v)
plt.plot(u, w)

plt.show()
```
